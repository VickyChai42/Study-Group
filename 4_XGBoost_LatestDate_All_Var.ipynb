```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train_final = X_train[["D_48", "B_18","B_2","B_9","D_44","D_75","D_55","B_33","D_58","B_7","B_3","B_23","D_74","B_4","B_16","B_1","B_19","B_37","B_20","B_22","B_11","R_1","D_62","D_42","D_77","S_7","B_8","S_3","D_61","D_70","R_3", "D_43","D_47","R_27", "B_17","D_45","D_78","R_2"
]]
y_train = train_data_final[["target"]]

model = XGBClassifier()
model.fit(X_train_final, y_train)
```

```python
test_data = pd.read_feather("test_data.ftr")
```

```python
test_data_1 = test_data[["customer_ID", "S_2", "D_48", "B_18","B_2","B_9","D_44","D_75","D_55","B_33","D_58","B_7","B_3","B_23","D_74","B_4","B_16","B_1","B_19","B_37","B_20","B_22","B_11","R_1","D_62","D_42","D_77","S_7","B_8","S_3","D_61","D_70","R_3", "D_43","D_47","R_27", "B_17","D_45","D_78","R_2"]].sort_values(['customer_ID', 'S_2'], ascending = [True, False])
test_data_2 = test_data_1.drop_duplicates(subset=["customer_ID"], keep="first")
test_data_2.isnull().sum()

```

```python
test_out_1 = model.predict_proba(test_data_2[["D_48", "B_18","B_2","B_9","D_44","D_75","D_55","B_33","D_58","B_7","B_3","B_23","D_74","B_4","B_16","B_1","B_19","B_37","B_20","B_22","B_11","R_1","D_62","D_42","D_77","S_7","B_8","S_3","D_61","D_70","R_3", "D_43","D_47","R_27", "B_17","D_45","D_78","R_2"]])
test_out_1
```

```python
test_out_2 = pd.DataFrame(test_out_1,columns=["1_minus_pred","prediction"])
test_out_2.head()
```

```python
X_test_2 = test_data_2[["customer_ID"]]

cus = X_test_2["customer_ID"].reset_index()
pred = test_out_2["prediction"]

test_out_3 = pd.concat([cus["customer_ID"],pred],ignore_index=True,axis=1)
test_out_3.rename(columns = {0:'customer_ID', 1:'prediction'}, inplace = True)

test_out_3.head()
```

```python
test_out_3.to_csv("Submission_XGboost_LatestDate_All_Var.csv", index=False)
```
